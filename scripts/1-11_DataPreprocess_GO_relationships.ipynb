```plaintext
library(GOSemSim)
library(dplyr)
library(GO.db)
```

```plaintext
data("gotbl", package = "GOSemSim")

all_relationships <- gotbl %>%
    dplyr::select(parent, go_id, relationship) %>%
    dplyr::left_join(term_info, by = c("parent" = "GOID")) %>%
    dplyr::rename(parent_term = Term, parent_ontology = Ontology) %>%
    dplyr::left_join(term_info, by = c("go_id" = "GOID")) %>%
    dplyr::rename(child_term = Term, child_ontology = Ontology)

write.csv(all_relationships, "../preprocessed_data/go_all_relationships.csv", row.names=FALSE)
```
