```python
import pandas as pd
import pickle
```

```python
HTMDT = pd.read_csv('../preprocessed_data/Preprocessed_HTMDT.csv')
herb_target = pd.read_csv('../preprocessed_data/Preprocessed_Herb-Target.csv')
```

```python
HTMDT
```

```python
herb1 = HTMDT[HTMDT['node1_type'] == 'Herb']['node1'].unique().tolist()
print(len(herb1))
herb2 = herb_target['Herb'].unique().tolist()
print(len(herb2))

herb = sorted(list(set(herb1) | set(herb2)))
print(len(herb))

herb_dict = {key: value for value, key in enumerate(herb)}
```

```python
TCMSymptom1 = HTMDT[HTMDT['node1_type'] == 'TCMSymptom']['node1'].unique().tolist()
TCMSymptom2 = HTMDT[HTMDT['node2_type'] == 'TCMSymptom']['node2'].unique().tolist()
TCMSymptom = sorted(list(set(TCMSymptom1) | set(TCMSymptom2)))
print(len(TCMSymptom))

TCMSymptom_dict = {key: value for value, key in enumerate(TCMSymptom)}
```

```python
MMSymptom1 = HTMDT[HTMDT['node1_type'] == 'MMSymptom']['node1'].unique().tolist()
MMSymptom2 = HTMDT[HTMDT['node2_type'] == 'MMSymptom']['node2'].unique().tolist()
MMSymptom = sorted(list(set(MMSymptom1) | set(MMSymptom2)))
print(len(MMSymptom))

MMSymptom_dict = {key: value for value, key in enumerate(MMSymptom)}
```

```python
Disease1 = HTMDT[HTMDT['node1_type'] == 'Disease']['node1'].unique().tolist()
Disease2 = HTMDT[HTMDT['node2_type'] == 'Disease']['node2'].unique().tolist()
Disease = sorted(list(set(Disease1) | set(Disease2)))
print(len(Disease))

Disease_dict = {key: value for value, key in enumerate(Disease)}
```

```python
with open('../preprocessed_data/Herb_dict.pkl', 'wb') as file:
    pickle.dump(herb_dict, file)
with open('../preprocessed_data/TCMSymptom_dict.pkl', 'wb') as file:
    pickle.dump(TCMSymptom_dict, file)
with open('../preprocessed_data/MMSymptom_dict.pkl', 'wb') as file:
    pickle.dump(MMSymptom_dict, file)
with open('../preprocessed_data/Disease_dict.pkl', 'wb') as file:
    pickle.dump(Disease_dict, file)
```
