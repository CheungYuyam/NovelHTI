```python
library(org.Hs.eg.db)
```

```python
file_path <- "../rawdata/DataS1_interactome.tsv"
data <- read.table(file_path, header = FALSE, sep = "\t", stringsAsFactors = FALSE)

entrez_ids_1 <- data$V1
entrez_ids_2 <- data$V2

entrez_ids_1 <- as.character(entrez_ids_1)
entrez_ids_2 <- as.character(entrez_ids_2)

entrez_ids_1 <- entrez_ids_1[!is.na(entrez_ids_1)]
entrez_ids_2 <- entrez_ids_2[!is.na(entrez_ids_2)]

convert_entrez_to_symbol <- function(entrez_ids) {
  symbols <- mget(entrez_ids, org.Hs.egSYMBOL, ifnotfound = NA)
  return(unlist(symbols))
}

gene_symbols_1 <- convert_entrez_to_symbol(entrez_ids_1)
gene_symbols_2 <- convert_entrez_to_symbol(entrez_ids_2)
```

```python
data$gene_symbol_1 <- gene_symbols_1
data$gene_symbol_2 <- gene_symbols_2
```

```python
write.table(data, "../preprocessed_data/converted_file.tsv", sep = "\t", row.names = FALSE, quote = FALSE)
```
