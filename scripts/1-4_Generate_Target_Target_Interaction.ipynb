```python
ppis = pd.read_csv('../preprocessed_data/converted_file.tsv', sep='\t')[['gene_symbol_1', 'gene_symbol_2', 'V3']]
ppis.columns = ['node1', 'node2', 'interaction']
ppis = ppis.dropna()

ppis_expanded = ppis.assign(interaction=ppis['interaction'].str.split(';')).explode('interaction')
protein_list = list(set(ppis_expanded['node1'].unique().tolist() + ppis_expanded['node2'].unique().tolist()))
```
