```python
import pandas as pd
import numpy as np
```

```python
HTMDT = pd.read_csv('../preprocessed_data/Preprocessed_HTMDT.csv', index_col=None)
HTMDT = HTMDT[(HTMDT['node1_type'] == 'Herb') & (HTMDT['node2_type'] == 'TCMSymptom')]
print(len(HTMDT))
herb_dict = pd.read_pickle('../preprocessed_data/herb_dict.pkl')
TCMSymptom_dict = pd.read_pickle('../preprocessed_data/TCMSymptom_dict.pkl')
HTMDT['node1'] = HTMDT['node1'].map(herb_dict)
HTMDT['node2'] = HTMDT['node2'].map(TCMSymptom_dict)
print(len(HTMDT))

Herb_TCM_matrix = np.zeros((698, 2311))
indices = HTMDT[['node1', 'node2']].values - 1
Herb_TCM_matrix[indices[:,0], indices[:,1]] = 1

print(Herb_TCM_matrix.shape)
```

```python
HTMDT = pd.read_csv('../preprocessed_data/Preprocessed_HTMDT.csv', index_col=None)
HTMDT = HTMDT[(HTMDT['node1_type'] == 'TCMSymptom') & (HTMDT['node2_type'] == 'MMSymptom')]
print(len(HTMDT))
TCMSymptom_dict = pd.read_pickle('../preprocessed_data/TCMSymptom_dict.pkl')
MMSymptom_dict = pd.read_pickle('../preprocessed_data/MMSymptom_dict.pkl')
HTMDT['node1'] = HTMDT['node1'].map(TCMSymptom_dict)
HTMDT['node2'] = HTMDT['node2'].map(MMSymptom_dict)
print(len(HTMDT))

TCM_MM_matrix = np.zeros((2311, 1143))
indices = HTMDT[['node1', 'node2']].values - 1
TCM_MM_matrix[indices[:,0], indices[:,1]] = 1

print(TCM_MM_matrix.shape)
```

```python
HTMDT = pd.read_csv('../preprocessed_data/Preprocessed_HTMDT.csv', index_col=None)
HTMDT = HTMDT[(HTMDT['node1_type'] == 'MMSymptom') & (HTMDT['node2_type'] == 'Disease')]
print(len(HTMDT))
MMSymptom_dict = pd.read_pickle('../preprocessed_data/MMSymptom_dict.pkl')
Disease_dict = pd.read_pickle('../preprocessed_data/Disease_dict.pkl')
HTMDT['node1'] = HTMDT['node1'].map(MMSymptom_dict)
HTMDT['node2'] = HTMDT['node2'].map(Disease_dict)
print(len(HTMDT))

MM_Disease_matrix = np.zeros((1143, 13992))
indices = HTMDT[['node1', 'node2']].values - 1
MM_Disease_matrix[indices[:,0], indices[:,1]] = 1

print(MM_Disease_matrix.shape)
```

```python
HTMDT = pd.read_csv('../preprocessed_data/Preprocessed_HTMDT.csv', index_col=None)
HTMDT = HTMDT[(HTMDT['node1_type'] == 'Disease') & (HTMDT['node2_type'] == 'Target')]
print(len(HTMDT))
Disease_dict = pd.read_pickle('../preprocessed_data/Disease_dict.pkl')
Target_dict = pd.read_pickle('../preprocessed_data/Target_dict.pkl')
HTMDT['node1'] = HTMDT['node1'].map(Disease_dict)
HTMDT['node2'] = HTMDT['node2'].map(Target_dict)
print(len(HTMDT))

Disease_Target_matrix = np.zeros((13992, 7854))
indices = HTMDT[['node1', 'node2']].values - 1
Disease_Target_matrix[indices[:,0], indices[:,1]] = 1

print(Disease_Target_matrix.shape)
```

```python
from scipy.sparse import csr_matrix, save_npz

Herb_TCM_sparse = csr_matrix(Herb_TCM_matrix)
TCM_MM_sparse = csr_matrix(TCM_MM_matrix)
MM_Disease_sparse = csr_matrix(MM_Disease_matrix)
Disease_Target_sparse = csr_matrix(Disease_Target_matrix)

save_npz('../preprocessed_data/Matrix_Herb_TCM.npz', Herb_TCM_sparse)
save_npz('../preprocessed_data/Matrix_TCM_MM.npz', TCM_MM_sparse)
save_npz('../preprocessed_data/Matrix_MM_Disease.npz', MM_Disease_sparse)
save_npz('../preprocessed_data/Matrix_Disease_Target.npz', Disease_Target_sparse)
```
