```python
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
df_final = pd.DataFrame(columns=['Target id', 'Gene symbol', 'Chromosome', 'Gene name', 'Protein name',
                                 'Ensembl id', 'UniProt id', 'P_value', 'FDR(BH)', 'FDR(Bonferroni)',
                                 'Relationship', 'Inferred evidence score', 'Herb'])

for i in tqdm(range(703)):

    rrid = f"SMHB{i+1:05d}"

    try:
        data = pd.read_csv(f'../rawdata/Herb-Target/{rrid}-Target.csv')
        data = data[data['P_value'] != 'None']
        data = data[data['Inferred evidence score'] != 'None']
        data['Herb'] = rrid

        if not df_final.empty:
            df_final = pd.concat([df_final, data], axis=0)
        else:
            df_final = data
    except:
        pass

df_final_protein = df_final.copy()

df_final = df_final[['Target id', 'Gene symbol', 'P_value', 'FDR(BH)', 
                     'FDR(Bonferroni)', 'Inferred evidence score', 'Herb']]

df_final = df_final.dropna()
df_final = df_final[df_final['Gene symbol'] != 'None']
df_final
```

```python
df_final_protein = df_final_protein[['Gene symbol', 'UniProt id', 'Ensembl id', 'Protein name']]
df_final_protein.to_csv('../preprocessed_data/Target_map.csv', index=None)
```

```python
df_final.to_csv('../preprocessed_data/Herb-Target.csv', index=None)
Target_list = pd.DataFrame(df_final['Gene symbol'].unique().tolist(), columns=['Proteins'])
Target_list.to_csv('../preprocessed_data/Target_list.csv', index=None)
```
